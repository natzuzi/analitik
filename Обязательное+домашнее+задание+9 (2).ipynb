```python
 #Выведите авторов, имеющихся в базе
show_all = """
SELECT *
FROM author;
"""

pd.read_sql(show_all, con)
```

```python
#Найдите всех авторов с инициалами "С. А.". Автор в базе определяется так: И. О. Фамилия
#корректный запрос 
request_3 = """
SELECT author.name
       
FROM author
WHERE author.name like 'С. А.%' 
"""


pd.read_sql(request_3, con)

```

```python
#Выведите список книг, выпущенных в издательстве "Мир".
request_1 = """
SELECT book.title, 
       publisher.name
FROM book, publisher  
WHERE (publisher.id = book.publisher_id) AND (publisher.name = 'Мир')
""" 
pd.read_sql(request_1, con)
```

```python
q = """
SELECT name FROM sqlite_master
WHERE type='table'
ORDER BY name;
"""
pd.read_sql(q,con)
```

```python
#Выведите количество книг, принадлежащих каждой из категорий. Результат отсортировать по возрастанию.
request_1 = """
SELECT  
        category.title AS title_category,
        COUNT(book.id) AS count_books
FROM  book, category 
WHERE (category.id = book.category_id)
GROUP BY category.title
ORDER BY COUNT(book.id) ASC              
           
"""
pd.read_sql(request_1,con)
```

```python
#Выведите количество книг,выпущенных в каждом издательстве. 
#В итоговом списке оставьте только те издательства, 
#которые выпустили 3 и более книг. Результат отсортировать по убыванию.
```

```python
get_books_publisher = """
    SELECT
        p.name as name_publishers,
        COUNT(b.title) as count_books
    FROM book b
        JOIN publisher p ON b.publisher_id = p.id
    GROUP BY 1
    HAVING count_books >= 3
    ORDER BY 2 DESC;
"""
df = pd.read_sql(get_books_publisher, con = con); df
```

```python

#Выведите список книг, автором которых являлся <<С. А. Айвазян>>

request_2 = """
SELECT author.name,
       book.title  
FROM book
INNER JOIN authorbook
ON book.id = authorbook.book_id
INNER JOIN author
ON author.id = authorbook.author_id
WHERE author.name = 'С. А. Айвазян'
"""
pd.read_sql(request_2, con)

```

```python
# В каких издательствах и сколько выпускались книги за авторством <<С. А. Айвазян>>. Результат отсортировать по убыванию.
request_3 = """
SELECT
    p.name as name_publisher,
    a.name as name_authors,
    COUNT(b.title) as count_books
FROM book b
         JOIN authorbook ab ON b.id = ab.book_id
         JOIN author a ON a.id = ab.author_id
         JOIN publisher p ON b.publisher_id = p.id
WHERE a.name = 'С. А. Айвазян'
GROUP BY 1
ORDER BY 3 DESC;
"""
df = pd.read_sql(request_3, con = con); df
```
