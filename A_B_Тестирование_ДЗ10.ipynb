H_0 : нет статистически значимых отличий в количестве пользователей, совершивших каждое событие по группам 246 и 248 a=b

H_0 : разница между количеством уникальных клиентов по каждой группе, совершивших определенное событие, статистически не значима a-b=0


H_1 : имеются статистически значимые отличия в количестве пользователей, совершивших каждое событие по группам 246 и 248


Смена цвета на зеленый приведет к увеличению показателя выручки на 48%.

Н_0: нет статистически значиных отличий в накопленных объемах выручки по группах A и B
H_1:

Для биноминальных распределений

$$
\begin{multline*}
z = \dfrac{\dfrac{m_1}{n_1}+ \dfrac{1}{2n_1} - \dfrac{m_2}{n_2} - \dfrac{1}{2n_2}}{\sqrt{\dfrac{m_1+m_2}{n_1+n_2} \dfrac{n_1+n_2 - m_1 - m_2}{n_1+n_2}\left(\dfrac{1}{n_1}+\dfrac{1}{n_2}\right)}} = \\
= \dfrac{\dfrac{m_1}{n_1}- \dfrac{m_2}{n_2} + \dfrac{1}{2n_1} - \dfrac{1}{2n_2}}{\sqrt{\dfrac{m_1+m_2}{n_1+n_2} \left( 1-\dfrac{m_1+m_2}{n_1+n_2}\right)\left(\dfrac{1}{n_1}+\dfrac{1}{n_2}\right)}}
\end{multline*}
$$

```python
# Проверьте гипотезу о наличие значимых отличий по результатам теста.
import math
import scipy.stats as st
```

```python
def z_test(group1, group2, event, alpha):
    p1_ev = df_all_groups.loc[event, group1]
    p2_ev = df_all_groups.loc[event, group2]
    p1_us = count_users.loc['247'][0]
    p2_us = count_users.loc['248'][0]
    p1 = p1_ev / p1_us
    p2 = p2_ev / p2_us
    difference = p1 - p2
    p_combined = (p1_ev + p2_ev) / (p1_us + p2_us)
    z_value = difference / math.sqrt(p_combined * (1 - p_combined) * (1 / p1_us + 1 / p2_us))
    distr = st.norm(0, 1)
    p_value = (1 - distr.cdf(abs(z_value))) * 2
    print('Группы {} и {}. Событие -- {}. P-значение --  {:.2f}. Уровень значимости -- {}'.format(group1, group2, event, p_value, alpha ))
    if (p_value < alpha):
        print("Отвергаем нулевую гипотезу -- между долями есть значимая разница")
    else:
        print("Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными")
```

```python
for event in df_all_groups.index:
    z_test('count_users_in_247', 'count_users_in_248', event, 0.05)
```
